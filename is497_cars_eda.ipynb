```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
df_original = pd.read_csv("cars_dataset_original.csv")
```

```python
df_original
```

```python
df = df_original.iloc[:, [13,14,15,8,10,6,3,16,17,1,0,2]]
```

```python
df = df.rename(columns={"Identification.Make":"Make",
                        "Identification.Model Year": "Model",
                        "Identification.Year": "Year",
                        "Engine Information.Number of Forward Gears":"Gears",
                        "Engine Information.Driveline":"Drive Type",
                        "Fuel Information.City mpg": "MPG City",
                        "Fuel Information.Highway mpg":"MPG Highway",
                        "Engine Information.Engine Statistics.Horsepower":"Horsepower",
                        "Engine Information.Engine Statistics.Torque":"Torque",
                        "Dimensions.Length":"Length",
                        "Dimensions.Height":"Height",
                        "Dimensions.Width":"Width"})
```

```python
df
```

```python
df['Make'].unique()
```

```python
df['Make'].replace({'Mercedes':"Mercedes-Benz",
                    'Mercedes-AMG': "Mercedes-Benz",
                    'AMG': "Mercedes-Benz",
                    'BMW Motorrad': "BMW",
                    'Grand Cherokee': "Jeep",
                    'Chrysler Group LLC': "Chrysler"
                   },
                  inplace=True)
```

```python
df['Make'].unique()
```

```python
df['Drive Type'].unique()
```

```python
df['Drive Type'].replace({'All-wheel drive': "AWD",
                          'Front-wheel drive': "FWD",
                          'Four-wheel drive':"AWD",
                          'Rear-wheel drive':"RWD"
                         },
                        inplace=True)
```

```python
df
```

```python
df['Length'].unique().sort()
```

```python
df.drop_duplicates(subset="Model",inplace=True)
```

```python
df.sort_values(by=["Make","Model"],ascending=True,inplace=True)
```

```python
df.set_index(["Make","Model"],inplace=True)
```

```python
df
```

```python
df.to_csv('cars_df_cleaned.csv')
```
